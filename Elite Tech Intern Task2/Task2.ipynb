# Unemployment Analysis using Python.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("unemployment_post_covid.csv")
```

```python
# Display basic info
print("Dataset Info:")
print(df.info())
print("\nFirst 5 Rows:")
print(df.head())
```

```python
df.dropna(inplace=True)  # Remove missing values
```

```python
# Plot Unemployment Rate Over Time
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x="Date", y="Unemployment_Rate", marker="o", hue="State", palette="tab10", linewidth=2.5)
plt.title("Unemployment Rate Over Time (Post-COVID)", fontsize=14, fontweight='bold')
plt.xlabel("Date", fontsize=12)
plt.ylabel("Unemployment Rate (%)", fontsize=12)
plt.xticks(rotation=45)
plt.legend(title="State", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle="--", alpha=0.7)
plt.show()
```

```python
# Bar Chart: Average Unemployment Rate by State
plt.figure(figsize=(10, 5))
sns.barplot(data=df, x="State", y="Unemployment_Rate", estimator=lambda x: sum(x)/len(x), hue="State", legend=False, palette="viridis")
plt.title("Average Unemployment Rate by State (2021-2024)", fontsize=14, fontweight='bold')
plt.xlabel("State", fontsize=12)
plt.ylabel("Avg. Unemployment Rate (%)", fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle="--", alpha=0.7)
plt.show()
```

```python
# Histogram to show the distribution of Unemployment Rate
plt.figure(figsize=(8, 5))
sns.histplot(df["Unemployment_Rate"], bins=20, kde=True, color="darkblue", alpha=0.8)
plt.title("Distribution of Unemployment Rate (Post-COVID)", fontsize=14, fontweight='bold')
plt.xlabel("Unemployment Rate (%)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(axis='y', linestyle="--", alpha=0.7)
plt.show()
```

```python
# Save the processed dataset
df.to_csv("cleaned_unemployment_post_covid.csv", index=False)
print("Processed dataset saved as 'cleaned_unemployment_post_covid.csv'")
```
